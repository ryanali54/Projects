##### 1.2.2) Scrapping Data from skytrax.com and writing them into lists

```python
driver.get("https://www.airlinequality.com/airline-reviews/british-airways/page/1/?pagesize=100&filterby=")
soup=BeautifulSoup(driver.page_source, "html.parser")
b=soup.find(class_="review-count")
b=b.text
b=b.replace("\t", "")
b=b.replace("\n", "")
b=b.split(" ")
total_reviews=int(b[3][0:4])
pages=total_reviews//100
review=[]
reviewer=[]
reviewtime=[]
review_heading=[]
masterrating=[]
place=[]
for z in range(1,pages+2):
    url=f"https://www.airlinequality.com/airline-reviews/british-airways/page/{z}/?pagesize=100&filterby="
    driver.get(url)
    driver.execute_script("window.scrollBy(0,document.body.scrollHeight)")
    soup=BeautifulSoup(driver.page_source, "html.parser")
    a=soup.find_all("h2", {"class": "text_header"})
    for i in range(len(a)):
        b=a[i].text
        b=b.replace('"', "")
        review_heading.append(b)
    a=soup.find_all("div", {"class": "text_content"})
    for i in range(len(a)):
        try:
            b=a[i].text
            b=b.split("|")
            c=b[1]
            review.append(c[2:])
        except IndexError:
            b=a[i].text
            review.append(b)
    a=soup.find_all("span", {"itemprop": "author"})
    for i in range(len(a)):
        b=a[i].text
        b=b.replace("\n","")
        reviewer.append(b)
    a=soup.find_all("h3",{"class":"text_sub_header userStatusWrapper"})
    b=[]
    for i in a:
        b.append(i.text)
    c=[]
    for i in b:
        c.append(i[4:])
    for i in c:
        e=i.find("(")
        f=i.find(")")
        place.append(i[e+1:f])
    
    print(f"Pass {z} Completed")
```

### 1.3) Sentiment Analysis using VaderSentimentAnalyzer

#### 1.3.1) Downloading required packages

#### 1.3.2) Cleaning the text (Removing all punctuation marks)

```python
def clean(text):
    text = re.sub('[^A-Za-z]+', ' ', str(text))
    return text
df['Cleaned Reviews'] = df['Review'].apply(clean)
df.head()
```

#### 1.3.3) Classifying the left over words as adjective, verb, noun and adverb

```python
pos_dict = {'J':wordnet.ADJ, 'V':wordnet.VERB, 'N':wordnet.NOUN, 'R':wordnet.ADV}
def token_stop_pos(text):
    tags = pos_tag(word_tokenize(text))
    l = []
    for word, tag in tags:
        if word.lower() not in set(stopwords.words('english')):
          l.append(tuple([word, pos_dict.get(tag[0])]))
    return l 

df['POS tagged'] = df['Cleaned Reviews'].apply(token_stop_pos)
df.head()
```

#### 1.3.4) Lemmatizing the words

```python
wordnet_lemmatizer = WordNetLemmatizer()
def lemmatize(pos_data):
    lemma_rew = " "
    for word, pos in pos_data:
        if not pos:
            lemma = word
            lemma_rew = lemma_rew + " " + lemma
        else:
            lemma = wordnet_lemmatizer.lemmatize(word, pos=pos)
            lemma_rew = lemma_rew + " " + lemma
    return lemma_rew

df['Lemma'] = df['POS tagged'].apply(lemmatize)
df.head()
```

#### 1.3.5) Analyzing Sentiment

```python
analyzer = SentimentIntensityAnalyzer()
def vadersentimentanalysis(review):
    vs = analyzer.polarity_scores(review)
    return vs['compound']
df['Sentiment'] = df['Lemma'].apply(vadersentimentanalysis)
def vader_analysis(compound):
    if compound >= 0.5:
        return 'Positive'
    elif compound < 0 :
        return 'Negative'
    else:
        return 'Neutral'
df['Analysis'] = df['Sentiment'].apply(vader_analysis)
df.head()
```
